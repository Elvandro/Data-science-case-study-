# Exploratory Data Analysis

```python
# Importing the required packages

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
# Loading the data and creating it's dataframe

data_df = pd.read_csv('Ongoza Data Science Intern _ Analyst Case Study - Data.csv')
```

```python
data_df
```

```python
data_df.info() # Summary of the dataframe 
```

### Data Cleaning

```python
data_df.isnull().sum() # Count the number of NaN values in each column
```

```python
# Deleting redundant or unnecessary columns

del_col_list = ['found_name1_present','found_name2_present','found_name3_present','found_additional_cnt',
                'found_name2_age','found_name3_age','found_name2_gender','found_name3_gender','found_name2_born',
               'found_name3_born','found_name2_resides','found_name3_resides','found_name2_education',
                'found_name3_education']
data_df = data_df.drop(del_col_list, axis=1)
data_df.head()
```

### Descriptive Statistics 

```python
data_df.describe().transpose() # Summary of numerical values in the dataframe
```

#### Continous Variables : 

> For continous variables we will look at the measures of central tendecy and the spread of the variable. We can use box plot, violin plot or histograms for visualization. 

```python
# Analyzing the revenue from the past year. From this we can derive what the firms make in profits or losses. 

data_df[['fins_revenues_m1']].describe()
```

```python
print("The highest revenue earned by a firm from the past year was: ", data_df['fins_revenues_m1'].max())
print("The lowest revenue earned by a firm from the past year was: ", data_df['fins_revenues_m1'].min())
print("The average revenue earned by firms from the past year was: ", data_df['fins_revenues_m1'].mean())
```

```python
# Analyzing the revenue since founding. From this we can determine the growth of a firm and its future potential. 

data_df[['fins_revenues_sincefound']].describe()
```

```python
print("The firm with the highest revenue earned since it was founded had: ", data_df['fins_revenues_sincefound'].max())
print("The firm with the lowest revenue earned since it was founded had: ", data_df['fins_revenues_sincefound'].min())
print("The average revenue earned by firms since founding was: ", data_df['fins_revenues_sincefound'].mean())
```

```python
# Looking at the number of full time employees of a firm, it is possible to conceptualize the size of the firm.

data_df[['fins_ft_employees_m1']].describe()
```

```python
print("The firm with the most full time employees has", data_df['fins_ft_employees_m1'].max(), "employees.")
print("The firm with the least number of full time employees has", data_df['fins_ft_employees_m1'].min(), "employee.")
print("The average number of full time employees is", round(data_df['fins_ft_employees_m1'].mean()), "employees.")
print("The standard deviation of this variable is", round(data_df['fins_ft_employees_m1'].std()),"- Comparing this value to the average, we can tell that the spread  of full time employees is close to the average.")
```

#### Categorical Variables : 

> For categorical variables we will look at the distribution of each category. Bar and pie charts can be used for visualization. For categorical (type object) and boolean (type bool) features we can use the value_counts method. Lets have a look.

```python
profit_margins_count = data_df['fins_profit_m1'].value_counts()
profit_margins_count
```

```python
profit_margins_count.plot.bar(figsize=[10,5], title="Firms Profit margins")
```

```python
region_count = data_df['venture region'].value_counts()
region_count
```

```python
region_count.plot.pie(figsize=[10,10], autopct='%1.1f%%' ,shadow=True)
```

```python
venture_hq_count = data_df['info_venture_country_hq'].value_counts()
venture_hq_count
```

```python
venture_hq_count.plot.bar(figsize=[20,8], title="Firms headquarter countries")
```
